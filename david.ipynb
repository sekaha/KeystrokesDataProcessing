```python
def get_bin(value, n):
    if n == 0:
        print(value)
    else:
        get_bin(value + "0", n - 1)
        get_bin(value + "1", n - 1)

get_bin("", 3)
```

```python
def get_all_strings(string, n):
    if n == 0:
        print(string)
    else:
        for c in "abc":
            get_all_strings(string + c, n - 1)

get_all_strings("", 2)
```

```python
# All numbers 0 to 255 in hex. These are used for RGB colors in things like CSS #FFFFFF #000000
def get_all_hex(num, n):
    if n == 0:
        print(num)
    else:
        for c in "0123456789ABCDEF":
            get_all_hex(num + c, n - 1)

get_all_hex("", 2)
```

```python
def get_triangular_number(n):
    if n == 0:
        return 0
    return get_triangular_number(n-1)+n

get_triangular_number(4)
```

```python
# actually had to use this lol
def get_harmonic(n):
    if n == 1:
        return 1
    else:
        return 1/n + get_harmonic(n-1)

get_harmonic(5)
print(137/60)
```

```python
def get_inverse_square(n):
    if n == 0:
        return 1
    else:
        return (1/(2**n))+get_inverse_square(n-1)

get_inverse_square(9)
```
