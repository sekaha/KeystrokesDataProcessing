```python
from classifier import keyboard, classifier
from scipy.optimize import curve_fit
import numpy as np

tristroke_times = {}
tristroke_freq = {}
tristrokes = set()
kb = keyboard()
key_classifier = classifier(kb)

with open(f"nstrokes/trigrams.txt") as file:
    for l in file:
        layout, tristroke, freq, *times = l.split("\t")

        if tristroke not in tristrokes:
            tristrokes.add(tristroke)
            
        tristroke_freq[(layout, tristroke)] = int(freq)
        tristroke_times[(layout, tristroke)] = [
            list(map(int, t.strip()[1:-1].split(", "))) for t in times
        ]
```
